## Can you predict the required number of bikes for rental given information about the weather and time of the day?

You are provided hourly rental data along with weather data. For this competition, the training set is comprised of the first 20 days of each month, while the test set is the 21th to the end of the month. You must predict the total count of bikes rented during each hour covered by the test set, using only information available prior to the rental period.

## Data fields
* <b>ID</b> - an ID for this instance<br>
* <b>Date</b> - year-month-day<br>
* <b>Hour</b> - Hour of he day<br>
* <b>Temperature</b> - Temperature in Celsius<br>
* <b>Humidity</b> - %<br>
* <b>Windspeed</b> - m/s<br>
* <b>Visibility</b> - 10m<br>
* <b>Dew point temperature</b> - Celsius<br>
* <b>Solar radiation</b> - MJ/m2<br>
* <b>Rainfall</b> - mm<br>
* <b>Snowfall</b> - cm<br>
* <b>Seasons</b> - Winter, Spring, Summer, Autumn<br>
* <b>Holiday</b> - Holiday/No holiday<br>
* <b>Functional Day</b> - NoFunc(Non Functional Hours), Fun(Functional hours)<br>
* <b>y</b> - Rented Bike count (Target), Count of bikes rented at each hour<br>

## Helper Function


```python
#Loading Data function
def load_data(data_path,name):
    csv_path=os.path.join(path,name)
    return pd.read_csv(csv_path)
###########################################################################################################################
#Clean columns Function
def clean_columns(df):
    return df.rename(columns={'Temperature(�C)': 'Temperature', 'Dew point temperature(�C)': 'Dew point temperature'})
###########################################################################################################################
# Recorrect Datatypes Function
def recorrect_Datatypes(df):
    
    df['Date'] = pd.to_datetime(df['Date'],format='%d/%m/%Y')
    df['year']= df['Date'].apply(lambda x:x.year).astype(int)
    df['month']= df['Date'].dt.month_name()
    df['woy'] = df['Date'].apply(lambda x:x.weekofyear).astype(int)
    df['week_day'] = df['Date'].dt.day_name()
    df.drop(columns=["Date",],inplace=True,axis=1)
    df['Snowfall (cm)']=df['Snowfall (cm)'].astype(int)
    df['Rainfall(mm)']=df['Rainfall(mm)'].astype(int)
    df['Solar Radiation (MJ/m2)']=df['Solar Radiation (MJ/m2)'].astype(int)
    return df
###########################################################################################################################
# RMSLE function
def RMSLE(y_true, y_pred):
    return np.sqrt(mean_squared_log_error(y_true, y_pred))
```

### Handling Numerical Attributes

```python
df=Training_data.copy()
```

```python
#Handling Numerical Attributes
df['ideal'] = df[['Temperature', 'Wind speed (m/s)']].apply(lambda x: (0, 1)[x['Temperature'] > 27 and x['Wind speed (m/s)'] < 3], axis = 1)
df['Solar Radiation (MJ/m2)']= np.sqrt(df['Solar Radiation (MJ/m2)'].astype(float))
df['Rainfall(mm)'] = df.rolling(2, min_periods=1)['Rainfall(mm)'].mean()
df['label_day_night']=df['Hour'].apply(lambda x : 'Night' if (x >20 or x<5) else( 'Day'))
df['dryness'] = 1 / (df['Rainfall(mm)'] + 1)
    
df['Visibility (10m)'] = df['Visibility (10m)'].max() - df['Visibility (10m)']
Num_Training_data=df[[ 'dryness','Visibility (10m)','Temperature', 'Humidity(%)', 'Wind speed (m/s)',
        'Dew point temperature', 'Solar Radiation (MJ/m2)',
       'Rainfall(mm)']]
scaler=StandardScaler()
scaler.fit(Num_Training_data)
Num_Training_data=pd.DataFrame(scaler.transform(Num_Training_data),index=Num_Training_data.index,columns=Num_Training_data.columns)
Num_Training_data
```

### Handling Text & Categorical Attributes


```python
# Handling Text & Categorical Attributes

df['Snowfall (cm)'] = df.rolling(8, min_periods=1)['Snowfall (cm)'].mean()
df['snowing'] = (df['Snowfall (cm)'] > 0).astype(np.int)
df = df.drop('Snowfall (cm)', axis=1)
df['ideal'] = df[['Temperature', 'Wind speed (m/s)']].apply(lambda x: (0, 1)[x['Temperature'] > 27 and x['Wind speed (m/s)'] < 3], axis = 1)

Cat_Training_data=df[['ideal','Hour', 'week_day','snowing','Seasons', 'Holiday',
       'Functioning Day',  'month', 'woy','label_day_night']]


cat_incoder1=OneHotEncoder()
cat_1hot=cat_incoder1.fit_transform(Cat_Training_data[['Seasons']]).toarray()
Seasons=pd.DataFrame(cat_1hot,columns=cat_incoder1.categories_,index=Cat_Training_data.index)
    
    
cat_incoder2=OneHotEncoder()
week_day_1hot=cat_incoder2.fit_transform(Cat_Training_data[['week_day']]).toarray()
week_day=pd.DataFrame(week_day_1hot,columns=cat_incoder2.categories_,index=Cat_Training_data.index)
    
cat_incoder3=OneHotEncoder()
month_1hot=cat_incoder3.fit_transform(Cat_Training_data[['month']]).toarray()
month=pd.DataFrame(month_1hot,columns=cat_incoder3.categories_,index=Cat_Training_data.index)
    
cat_incoder4=OneHotEncoder()
Hour_1hot=cat_incoder4.fit_transform(Cat_Training_data[['Hour']]).toarray()
Hour=pd.DataFrame(Hour_1hot,columns=cat_incoder4.categories_,index=Cat_Training_data.index)
    
    
    
ordinal_encoder=OrdinalEncoder()
Ord_Encoded=ordinal_encoder.fit_transform(Cat_Training_data[['ideal','snowing','Holiday','Functioning Day','label_day_night']])
Cat_1=pd.DataFrame(Ord_Encoded,columns=['ideal','snowing','Holiday','Functioning Day','label_day_night'],index=Cat_Training_data.index)

Cat_Training_data=pd.concat([Cat_1,Seasons,week_day,month],axis=1)
df=df.drop(columns=['ideal','month','week_day','dryness','snowing','ID','Temperature', 'Humidity(%)', 'Wind speed (m/s)',
       'Dew point temperature', 'Solar Radiation (MJ/m2)',
       'Rainfall(mm)','Seasons', 'Holiday',
       'Functioning Day', 'Visibility (10m)','year', 'month', 'woy','label_day_night'])

Training_data=pd.concat([df,Hour,Cat_Training_data,Num_Training_data],axis=1)
```


# Model

```python
y_log = np.log1p(Training_data['y'])
X_train, X_test, y_train, y_test= train_test_split(Training_data.drop(columns=[ 'y']),y_log, test_size=.5,shuffle=True, random_state=42)
```

```python
# Define the grid of hyperparameters to search

params = { 'max_depth': [3, 5, 6, 10, 15, 20],
           'learning_rate': [0.01, 0.1, 0.2, 0.3],
           'subsample': np.arange(0.5, 1.0, 0.1),
           'colsample_bytree': np.arange(0.3, 1.0, 0.1),
           'colsample_bylevel': np.arange(0.4, 1.0, 0.1),
           'n_estimators': [100, 500, 1000]
         }
xgbr = xgb.XGBRegressor(seed = 20)
reg = RandomizedSearchCV(estimator=xgbr,param_distributions=params,
                         scoring='neg_mean_squared_error',
                         n_iter=35,cv=5,
                         verbose=1,
                         return_train_score = True,
            random_state=42)
reg.fit(X_train, y_train)
print("Best parameters:", reg.best_params_)
print("Lowest RMSE: ", (-reg.best_score_)**(1/2.0))
```

```python
y_pred=reg.predict(X_test)
```

```python
x_ax = range(len(np.exp(y_test)))
plt.figure(figsize=(20,10))

plt.hist( np.exp(y_test),bins=30, label="original")

plt.hist(np.exp(y_pred),alpha=.9,bins=30, label="predicted")
plt.title("Test and predicted data")
plt.legend()
plt.show()
```
